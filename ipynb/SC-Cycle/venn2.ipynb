```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

plt.style.use('default')
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
plt.rcParams['font.family'] = 'Arial'

report_dir = Path('../../reports/SC-Cycle')

df = pd.read_parquet(report_dir/'diann_220'/'report.parquet', engine='fastparquet')
prs_diann = set(df[df['Global.Q.Value'] < 0.01]['Precursor.Id'])
pgs_diann = set(df[df['Global.PG.Q.Value'] < 0.01]['Protein.Group'])

df = pd.read_parquet(report_dir/'full_100'/'report.parquet')
prs_full = set(df[df['Global.Q.Value'] < 0.01]['Precursor.Id'])
pgs_full = set(df[(df['Global.PG.Q.Value'] < 0.01)]['Protein.Group'])
```

```python
v = venn2([prs_diann, prs_full], ['DIA-NN', 'Full-DIA'])
for t in v.set_labels + v.subset_labels:
    if t: t.set_fontsize(18)
plt.tight_layout()
plt.show()
```

```python
v = venn2([pgs_diann, pgs_full], ['DIA-NN', 'Full-DIA'])
for t in v.set_labels + v.subset_labels:
    if t: t.set_fontsize(18)
plt.tight_layout()
plt.show()
```
